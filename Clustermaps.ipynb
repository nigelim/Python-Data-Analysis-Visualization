```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas import Series, DataFrame

df = sns.load_dataset('flights')
df2 = df.pivot('year','month','passengers')

#clustermap
sns.clustermap(df2).savefig('cluster1.png')
#-> basically the data is clustered based on their values
```

```python
#Create cluster along one direction (axis)
plt.clf()
sns.clustermap(df2,col_cluster=False).savefig('cluster2.png')
```

```python
#STANDARDISE
#standard_scale=0 : standardise by rows,
plt.clf()
sns.clustermap(df2,standard_scale=0).savefig('cluster3.png')
```

```python
plt.clf()
# standard_scale = 1 : standardise by columns
sns.clustermap(df2,standard_scale=1).savefig('cluster4.png')
# standard_scale = 1 : standardise by columns
```
