```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas import Series, DataFrame

ds1= randn(1000)
ds2 = randn(100)

plt.hist(ds1)
plt.savefig('image1.png')

plt.hist(ds2)
plt.savefig('image2.png')
```

```python
plt.hist(ds1,cumulative=True,color='green',bins=15, alpha=0.5)
plt.hist(ds2,cumulative=True,bins=15,alpha=0.5)
plt.savefig('image3.png')
```

```python
ds3 = randn(1000)
ds4 = randn(1000)

sns_plot = sns.jointplot(ds3)
sns_plot.savefig('image4.png')

sns_plot2 = sns.jointplot(ds3, kind="hex")
sns_plot2.savefig('image5.png')
plt.show()
```
