```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas import Series, DataFrame

#Violin plot is a combination of Box plot + KDE plot
plt.clf()
ds = randn(80)
sns.violinplot(ds, orient='h').get_figure().savefig('violin1.png')
```

```python
plt.clf()
sns.violinplot(ds, bw=0.2, orient='h').get_figure().savefig('violin2.png')
```

```python
plt.clf()
sns.violinplot(ds, inner='stick', orient='h').get_figure().savefig('violin3.png')
```
