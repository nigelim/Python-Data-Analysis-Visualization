```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas import Series, DataFrame

df = sns.load_dataset('diamonds').sample(frac=1).head(100) 
#->diamonds is a .csv file; can be found on GitHub(seaborn data)
print(df)
```

```python
plt.clf()
sns.lmplot(df, x='price',y='carat').savefig('Regress1.png')
```

```python
#modify
plt.clf()
sns.lmplot(df, x='price', y='carat',scatter_kws={'marker':'o','color':'green'},line_kws={'color':'red','linewidth':1}).savefig('Regress2.png')
```

```python
#higher order trendline (non-linear lines) using order =''
plt.clf()
sns.lmplot(df, order=4, x='price', y='carat',scatter_kws={'marker':'o','color':'green'},line_kws={'color':'red','linewidth':1}).savefig('Regress3.png')
```

```python
#scatter plot without a trendline
plt.clf()
sns.lmplot(df,fit_reg=False, x='price',y='carat').savefig('Regress4.png')
```

```python
#hue_arguments
plt.clf()
sns.lmplot(df, hue='cut', x='price',y='carat').savefig('Regress5.png')
```

```python
plt.clf()
sns.lmplot(df, hue='cut',markers=['^','v','*','.','s'], x='price',y='carat').savefig('Regress6.png')
```

```python
#local regression
plt.clf()
sns.lmplot(df, lowess=True, x='price',y='carat').savefig('Regress7.png')
```

```python
#regplot
sns.regplot(df, x='price', y='carat').get_figure().savefig('RegPlot1.png')
```
